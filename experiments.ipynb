```python
import os
import torch
from model import GPT, GPTConfig

def get_model_by_path(ckpt_path, device):
    checkpoint = torch.load(ckpt_path, map_location=device)
    gptconf = GPTConfig(**checkpoint['model_args'])
    model = GPT(gptconf)
    state_dict = checkpoint['model']
    model.load_state_dict(state_dict)
    return model
```

```python
import pickle

def load_tokenizer(meta_path, ckpt_path, device="cpu"):
    checkpoint = torch.load(ckpt_path, map_location=device)
    meta_path = os.path.join('data', checkpoint['config']['dataset'], 'meta.pkl')

    print(f"Loading meta from {meta_path}...")
    with open(meta_path, 'rb') as f:
        meta = pickle.load(f)
    # TODO want to make this more general to arbitrary encoder/decoder schemes
    stoi, itos = meta['stoi'], meta['itos']
    encode = lambda s: [stoi[c] for c in s]
    decode = lambda l: ''.join([itos[i] for i in l])
    
    return encode, decode
```

```python
gpt2_mini_6l_5h_classic_mul = get_model_by_path("gpt2-mini-6l-5h-emb768-do-0-classic-mul.pt", "cpu")

encode, decode = load_tokenizer("data/digit_mul/meta.pkl", "gpt2-mini-6l-5h-emb768-do-0-classic-mul.pt")

start = (torch.tensor(encode("1111*1111"), dtype=torch.long, device="cpu")[None, ...])
y = gpt2_mini_6l_5h_classic_mul.generate(start, 100, stop=encode("\n")[0])
print(decode(y[0].tolist()), y[0].tolist().__len__())
s_in = encode("1111*1111 = [1111, 1111, 1111, 1111] => ")
start = (torch.tensor(s_in, dtype=torch.long, device="cpu")[None, ...])
out, _, hs = gpt2_mini_6l_5h_classic_mul.forward(start, return_hidden=True)

for i in range(6):
    gram = hs[i].squeeze() @ hs[i].squeeze().T
    lambdas, _ = torch.linalg.eig(gram)
    lambdas = torch.real(lambdas)
    alpha = 0.99
    entropy = 1 / (1-alpha) *torch.log(sum((lambdas / sum(lambdas))**alpha))
    print(entropy)
```

```python
def get_mean_entropy(model_path, test):
    ent_all = []
    model = get_model_by_path(model_path, "cpu")
    encode, decode = load_tokenizer("data/digit_mul/meta.pkl", model_path)
    for s in test:
        # start = (torch.tensor(encode(s), dtype=torch.long, device="cpu")[None, ...])
        # y = model.generate(start, 100, stop=encode("\n")[0])
        # print(decode(y[0].tolist()), y[0].tolist().__len__())

        s_in = encode(s)
        start = (torch.tensor(s_in, dtype=torch.long, device="cpu")[None, ...])
        out, _, hs = model.forward(start, return_hidden=True)
        ent = []
        for i in range(6):
            gram = hs[i].squeeze() @ hs[i].squeeze().T
            lambdas, _ = torch.linalg.eig(gram)
            lambdas = torch.real(lambdas)
            alpha = 0.99
            entropy = 1 / (1-alpha) *torch.log(sum((lambdas / sum(lambdas))**alpha))
            ent.append(entropy)
        ent_all += [torch.tensor(ent)]
    return ent_all

```

```python
import torch
import random
import tqdm

def get_samples_by_model(model_path, l, device="cpu"):
    model = get_model_by_path(model_path, device)
    encode, decode = load_tokenizer("data/digit_mul/meta.pkl", model_path)
    out = []
    for _ in tqdm.tqdm(range(l)):
        num1 = random.randint(1000, 9999)
        num2 = random.randint(1000, 9999)
        res = num1 * num2
        s = f"{num1}*{num2}"
        
        start = (torch.tensor(encode(s), dtype=torch.long, device=device)[None, ...])
        y = model.generate(start, 100, stop=encode("\n"))
        y = decode(y[0].tolist())
        try:
            y_pred = y[y.index("=> ") + 3:y.index("\n")]
            out.append([num1, num2, res, int(y_pred)])
        except:
            pass
        
    return out
```
