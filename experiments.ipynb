```python
import wandb

api = wandb.Api()
runs = api.runs(path="kurvadimal-astrakhan-state-university-of-architecture-an/reflex_attention")
run_histories = {}
for i, run in enumerate(runs):
  run_histories[run.name] = run._full_history()
  print("run name = ",run.name," id: ", run.id)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

l6_classic = run_histories["gpt2-mini-6l-5h-emb768-do-0-classic"][:10]
l6_reflex = run_histories["gpt2-mini-6l-5h-emb768-do-0-reflex"][:10]
l6_route = run_histories["gpt2-mini-6l-5h-emb768-do-0-route"][:10]

fig, axs = plt.subplots(2, 2, figsize=(10,6))
sns.set_style("darkgrid")
flatui = ["#95a5a6", "#34495e"]

sns.lineplot(x = [i for i in range(1, len(l6_classic)+1)], y=[x['val/loss'] for x in l6_classic], ax=axs[0, 0] , label='Classic attention')
sns.lineplot(x = [i for i in range(1, len(l6_reflex)+1)], y=[x['val/loss'] for x in l6_reflex], ax=axs[0, 0], label='Reflex attention')

axs[0,0].set_title("Val loss")
axs[0,0].set_ylabel("Loss")
axs[0,0].set_xlabel("Steps")

sns.lineplot(x = [i for i in range(1, len(l6_classic)+1)], y=[x['train/loss'] for x in l6_classic], ax=axs[0, 1], label='Classic attention')
sns.lineplot(x = [i for i in range(1, len(l6_reflex)+1)], y=[x['train/loss'] for x in l6_reflex], ax=axs[0, 1], label='Reflex attention')

axs[0, 1].set_title("Train loss")
axs[0, 1].set_ylabel("Loss")
axs[0, 1].set_xlabel("Steps")

sns.lineplot(x = [i for i in range(7, len(l6_classic)+1)], y=[x['val/loss'] for x in l6_classic[6:]], ax=axs[1, 0] , label='Classic attention')
sns.lineplot(x = [i for i in range(7, len(l6_reflex)+1)], y=[x['val/loss'] for x in l6_reflex[6:]], ax=axs[1, 0], label='Reflex attention')

axs[1,0].set_title("Zoomed Val loss")
axs[1,0].set_ylabel("Loss")
axs[1,0].set_xlabel("Steps")

sns.lineplot(x = [i for i in range(7, len(l6_classic)+1)], y=[x['train/loss'] for x in l6_classic[6:]], ax=axs[1, 1], label='Classic attention')
sns.lineplot(x = [i for i in range(7, len(l6_reflex)+1)], y=[x['train/loss'] for x in l6_reflex[6:]], ax=axs[1, 1], label='Reflex attention')

axs[1, 1].set_title("Zoomed Train loss")
axs[1, 1].set_ylabel("Loss")
axs[1, 1].set_xlabel("Steps")


plt.tight_layout()
plt.savefig('images/reflex_attn_compare.png')
```

```python
import os
import torch
from model import GPT, GPTConfig

def get_model_by_path(ckpt_path, device):
    checkpoint = torch.load(ckpt_path, map_location=device)
    gptconf = GPTConfig(**checkpoint['model_args'])
    model = GPT(gptconf)
    state_dict = checkpoint['model']
    # unwanted_prefix = '_orig_mod.'
    # for k,v in list(state_dict.items()):
    #     if k.startswith(unwanted_prefix):
    #         state_dict[k[len(unwanted_prefix):]] = state_dict.pop(k)
    model.load_state_dict(state_dict)
    return model
```

```python
gpt2_mini_6l_5h_classic = get_model_by_path("gpt2-mini-6l-5h-emb768-do-0-classic-ckpt.pt", "cuda")
```

```python
gpt2_mini_6l_5h_reflex= get_model_by_path("gpt2-mini-6l-5h-emb768-do-0-reflex-ckpt.pt", "cuda")
gpt2_mini_6l_5h_route = get_model_by_path("gpt2-mini-6l-5h-emb768-do-0-route-ckpt.pt", "cuda")
```

```python
H = gpt2_mini_6l_5h_route.config.n_head
HS = 3
v_router_values = []
for i in range(2,len(gpt2_mini_6l_5h_route.transformer.h)):
    v_router_values.append(gpt2_mini_6l_5h_route.transformer.h[i].attn.v_router.data.view(HS, H))
v_router_values = torch.stack(v_router_values)

k_router_values = []
for i in range(2,len(gpt2_mini_6l_5h_route.transformer.h)):
    k_router_values.append(gpt2_mini_6l_5h_route.transformer.h[i].attn.k_router.data.view(HS, H))
k_router_values = torch.stack(k_router_values)

v_router_values, k_router_values
```

```python
from string import ascii_letters
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl



# Compute the correlation matrix


# Set up the matplotlib figure
f, ax = plt.subplots(4,2,figsize=(12, 10))

cbar_ax = f.add_axes([.91, .3, .03, .4])

t = torch.stack([k_router_values, v_router_values])
vmin = t.min() 
vmax = t.max() 

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

f.suptitle("Values for $K_{router}$ and $V_{router}$", fontsize=18)

for i in range(k_router_values.shape[0]):
    corr = k_router_values[i]
    ax[i,0].set_title(f"$Layer_{i+3}$")
    # Draw the heatmap with the mask and correct aspect ratio
    sns.heatmap(corr, cmap=sns.color_palette("vlag", as_cmap=True), 
                vmin=vmin, vmax=vmax, 
                annot = True, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .5}, 
                ax=ax[i,0], cbar_ax=cbar_ax)

for i in range(v_router_values.shape[0]):
    corr = v_router_values[i]
    # Draw the heatmap with the mask and correct aspect ratio
    ax[i,1].set_title(f"$Layer_{i+3}$")
    sns.heatmap(corr, cmap=sns.color_palette("vlag", as_cmap=True), 
                vmin=vmin, vmax=vmax, 
                annot = True, center=0,
                square=True, linewidths=.5, cbar_kws={"shrink": .5}, 
                ax=ax[i,1], cbar_ax= cbar_ax)
    
for i in range(v_router_values.shape[0]):
    ax[i,0].set_yticklabels(["$Layer_{i-2}$", "$Layer_{i-1}$", "$Layer_{i}$"], rotation=0, ha='right')
    ax[i,1].set_yticks([])
    
    ax[i,0].set_xticklabels([])
    ax[i,1].set_xticklabels([])

for i in range(2):
    ax[-1,i].set_xticklabels([f"$h_{i}$" for i in range(1, H+1)], rotation=0, ha='right')


f.tight_layout(rect=[0, 0, .95, 1])

plt.savefig('images/layers_connection_compare.png')
```
